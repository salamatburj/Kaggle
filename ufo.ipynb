```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib notebook
import seaborn as sns
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.shape
```

```python
ufo['Shape Reported'].unique()
```

```python
ufo_sorted=ufo.groupby('Shape Reported').count().sort_values('City',ascending=False)
ufo_sorted.head()
```

```python
ufo_sorted.index
```

```python
plt.figure()
ind=np.arange(len(ufo_sorted.index))
plt.bar(ind, ufo_sorted.City)
plt.xticks(ind,ufo_sorted.index,rotation='vertical')

plt.ylabel('Number of reported observations')
plt.tight_layout()
```

```python
ufo_color=ufo.groupby('Colors Reported').count().sort_values('City',ascending=False)
```

```python
plt.figure()
ind=np.arange(len(ufo_color.index))
plt.bar(ind, ufo_color.City)
plt.xticks(ind,ufo_color.index,rotation='vertical')
plt.ylabel('Number of reported observations')
plt.tight_layout()
```

```python
ufo_state=ufo.groupby('State').count().sort_values('City',ascending=False)
```

```python
plt.figure()
ind=np.arange(len(ufo_state.index))
plt.bar(ind, ufo_state.City)
plt.xticks(ind,ufo_state.index,rotation='vertical')
plt.ylabel('Number of reported observations')
plt.xlim(-0.5,20.5)
plt.tight_layout()
```
